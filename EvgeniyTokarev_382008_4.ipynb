Импортирует нужные библиотеки, а именно numpy, matplotlib.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
%matplotlib inline
```

Задаем координаты своим будущим точкам и изображаем



```python
x = np.array([1, 4, 6, 4, 2, 1, 4, 7])
y = np.array([1, 0, 1, 3, 4, 6, 8, 6])
plt.plot(x, y, 'o', color='green')
pass
```

Создаём функцию для кубической интерполяции.

```python
def cubic_interpol():
  i = np.array([1, 2, 3, 4, 5, 6, 7, 8])
  t = np.linspace(1, 8, 500)
  f = np.polyfit(i, x, 3)
  g = np.polyfit(i, y, 3)
  xx = np.polyval(f, t)
  yy = np.polyval(g, t)
  plt.plot(x, y, 'o', color='green')
  plt.plot(xx, yy, color="tab:blue")
  pass
```

Создаём функцию для кубического сплайна.

```python
def cubic_spline(): 
  f = interpolate.interp1d(i, x, kind='cubic')
  g = interpolate.interp1d(i, y, kind='cubic')
  xx = f(t)
  yy = g(t)
  plt.plot(x, y, 'o', color='green')
  plt.plot(xx, yy, color="tab:blue")
  pass
```

Как можно заметить, линия с точками не соприкасается.

```python
cubic_interpol()
```

А здесь, наоборот, линии проходят сквозь точки.

```python
cubic_spline()
```

В итоге, хотелось бы сказать, что графики получается очень похожими, но кубическая интерполяция выдает очень плавный график, а сплайновая функция выдает более "точный" график.
